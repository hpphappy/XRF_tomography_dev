# Plot model MSE

```python
gt_path = "../data/sample8_size_64_pad/nElements_2/grid_concentration.npy"
gt_grid = np.load(gt_path)
print(gt_grid.shape)
```

```python
recon_folder = "../data/sample_8_size_64_test_recon_probeAtt_False"
recon_folder_probeAtt = "../data/sample_8_size_64_test_recon"
recon_folder_probeAtt_Ab = "../data/sample_8_size_64_test_recon_Ab"
recon_folder_probeAtt_XRT = "../data/sample_8_size_64_test_recon_XRFT_b1_10E4_b2_1E0"
recon_folder_probeAtt_Ab_XRT = "../data/sample_8_size_64_test_recon_Ab_XRFT_b1_10E4_b2_1E0"

recon_dir_ls = [recon_folder, recon_folder_probeAtt, recon_folder_probeAtt_Ab, recon_folder_probeAtt_XRT, recon_folder_probeAtt_Ab_XRT]
# recon_dir_ls = [recon_folder, recon_folder_probeAtt, recon_folder_probeAtt_XRT, recon_folder_probeAtt_Ab_XRT]
model_mse = np.zeros((5, 100))
for i, folder in enumerate(recon_dir_ls):
    for epoch in range(100):
        checkpoint = os.path.join(folder, f"checkpoint/grid_concentration_{epoch}.h5")
        with h5py.File(checkpoint, 'r') as s:
            recon_grid = s["sample/densities"][...]
        model_mse[i, epoch] = np.mean((recon_grid - gt_grid)**2)
```

```python
epoch = np.arange(100)
fig, ax = plt.subplots(figsize=(6,4), ncols=1, nrows=1)

ax_plot = ax.plot(epoch, model_mse[0], color='#AF1E1E', linewidth=4, alpha=0.5, label = 'No_probeAtt')
ax_plot = ax.plot(epoch, model_mse[1], color='#002E79', linewidth=2, alpha=1.0, label = 'probeAtt')
ax_plot = ax.plot(epoch, model_mse[2], color='#d62728', linewidth=2, alpha=1.0, label = 'probeAtt_Ab')
ax_plot = ax.plot(epoch, model_mse[3], linestyle=(0, (3, 2)), color='#2ca02c', linewidth=2, alpha=1.0, label = 'probeAtt_XRT')
ax_plot = ax.plot(epoch, model_mse[4], linestyle=(0, (3, 2)), color='#ff7f0e', linewidth=2, alpha=0.5, label = 'probeAtt_XRT_Ab')


# ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("i_epoch", fontsize=14)
ax.set_ylabel("model MSE", fontsize=14)

ax.tick_params(axis='both', which='major', labelsize=14)
ax.tick_params(axis='both', which='minor', labelsize=14)  
ax.legend(frameon=False, fontsize=14)
plt.savefig('model_convergence.pdf', format='pdf', dpi=300)
```
